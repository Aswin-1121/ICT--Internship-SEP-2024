```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/credit.csv')
data.head()
```

**Loading the 3 datasets**

```python
tr_data= pd.read_csv('/content/train_data.csv')
tr_data.head()
```

```python
ts_data= pd.read_csv('/content/test_data.csv')
ts_data.head()
```

```python
val_data= pd.read_csv('/content/val_data.csv')
val_data.head()
```

**Feature Engineering**

```python
# Create interaction features (e.g., multiplying two numerical features)
def create_interaction_features(data):
    data['Income_Utilization'] = data['Annual_Income'] * data['Credit_Utilization_Ratio']
    data['Debt_to_Income'] = data['Outstanding_Debt'] / (data['Annual_Income'] + 1)
    return data

# Apply to all datasets
train_data = create_interaction_features(tr_data)
val_data = create_interaction_features(val_data)
test_data = create_interaction_features(ts_data)
```

```python
 #Binning Age into categories
def bin_age(data):
    data['Age_Bin'] = pd.cut(data['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '30-40', '40-50', '50-60', '60+'])
    return data

# Apply to all datasets
train_data = bin_age(tr_data)
val_data = bin_age(val_data)
test_data = bin_age(ts_data)
```

```python
# Log transform to handle skewed data
def log_transform_skewed_features(data):
    data['Log_Outstanding_Debt'] = np.log1p(data['Outstanding_Debt'])
    data['Log_Annual_Income'] = np.log1p(data['Annual_Income'])
    return data

# Apply to all datasets
train_data = log_transform_skewed_features(tr_data)
val_data = log_transform_skewed_features(val_data)
test_data = log_transform_skewed_features(ts_data)
```

**One hot encoding**

```python
# Apply one-hot encoding to categorical variables (drop_first=True to avoid multicollinearity)
train_data_encoded = pd.get_dummies(train_data, drop_first=True)
val_data_encoded = pd.get_dummies(val_data, drop_first=True)
test_data_encoded = pd.get_dummies(test_data, drop_first=True)

# Ensure consistent columns across all datasets
common_cols = train_data_encoded.columns.intersection(val_data_encoded.columns).intersection(test_data_encoded.columns)

# Filter all datasets to keep only common columns
train_data_encoded = train_data_encoded[common_cols]
val_data_encoded = val_data_encoded[common_cols]
test_data_encoded = test_data_encoded[common_cols]
```

**Modelling**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Define features and target variable
X_train = train_data_encoded.drop(columns=['Credit_Score'])
y_train = train_data_encoded['Credit_Score']

X_val = val_data_encoded.drop(columns=['Credit_Score'])
y_val = val_data_encoded['Credit_Score']

X_test = test_data_encoded.drop(columns=['Credit_Score'])
y_test = test_data_encoded['Credit_Score']
```

```python
# Define models
models = {
    'KNN': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'SVM': SVC()
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Dictionary to store model performance
model_performance = {}

# Function to evaluate a model and store its performance
def evaluate_model(model, X_train, y_train, X_val, y_val, model_name):
    model.fit(X_train, y_train)  # Train the model
    y_pred_val = model.predict(X_val)  # Predictions on validation data

    # Compute metrics
    accuracy = accuracy_score(y_val, y_pred_val)
    precision = precision_score(y_val, y_pred_val, average='weighted')
    recall = recall_score(y_val, y_pred_val, average='weighted')
    f1 = f1_score(y_val, y_pred_val, average='weighted')
    conf_matrix = confusion_matrix(y_val, y_pred_val)
    class_report = classification_report(y_val, y_pred_val, output_dict=True)

    # Store metrics in dictionary
    model_performance[model_name] = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'confusion_matrix': conf_matrix,
        'classification_report': class_report
    }
```

```python
# Evaluate each model and store its performance
for model_name, model in models.items():
    evaluate_model(model, X_train, y_train, X_val, y_val, model_name)

# Display the performance for each model
for model_name, performance in model_performance.items():
    print(f"\nModel: {model_name}")
    print(f"Accuracy: {performance['accuracy']}")
    print(f"Precision: {performance['precision']}")
    print(f"Recall: {performance['recall']}")
    print(f"F1-Score: {performance['f1_score']}")
    print(f"Confusion Matrix:\n{performance['confusion_matrix']}")
    print(f"Classification Report: {performance['classification_report']}")
```

**Selecting best model**

```python
# Assuming 'Random Forest' performed best based on validation metrics
best_model_name = 'Random Forest'
best_model = models[best_model_name]

# Train the best model on full training data and evaluate on the test set
best_model.fit(X_train, y_train)
y_pred_test = best_model.predict(X_test)

# Calculate metrics on the test set
accuracy_test = accuracy_score(y_test, y_pred_test)
precision_test = precision_score(y_test, y_pred_test, average='weighted')
recall_test = recall_score(y_test, y_pred_test, average='weighted')
f1_test = f1_score(y_test, y_pred_test, average='weighted')
conf_matrix_test = confusion_matrix(y_test, y_pred_test)
```

```python
 #Print out the test metrics
print(f"\nTest Set Performance for {best_model_name}:")
print(f"Test Accuracy: {accuracy_test}")
print(f"Test Precision: {precision_test}")
print(f"Test Recall: {recall_test}")
print(f"Test F1-score: {f1_test}")
print(f"Test Confusion Matrix:\n{conf_matrix_test}")
```
