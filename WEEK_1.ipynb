Detecting Outliers

```python
numerical_columns = [
    'Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
    'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date',
    'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries',
    'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age',
    'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance'
]
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(20, 15))

# Plot box plots for each numerical column to visualize outliers
for i, col in enumerate(numerical_columns):
    plt.subplot(5, 4, i+1)
    sns.boxplot(x=data[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Remove outliers based on the IQR method for each column
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Apply outlier removal
data = remove_outliers(data, numerical_columns)

# Display the cleaned dataset summary
print(data.describe())
```

Encoding and scaling

```python
from sklearn.preprocessing import LabelEncoder

# List of columns that need label encoding
num_obj_col = ['Credit_Score', 'Payment_Behaviour', 'Payment_of_Min_Amount',
               'Credit_Mix', 'Type_of_Loan', 'Occupation', 'Name']

# Initialize the LabelEncoder
ln = LabelEncoder()

# Apply LabelEncoder to each specified column using map and lambda functions
data[num_obj_col] = data[num_obj_col].apply(lambda col: ln.fit_transform(col))

# Display the first few rows of the updated dataframe
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Function to scale the DataFrame
def scale_dataframe(df):
    # Initialize the StandardScaler
    scaler = StandardScaler()

    # Fit and transform the data, then convert it back to a DataFrame
    scaled_data = scaler.fit_transform(df)
    scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

    return scaled_df

# Apply the scaling function to the data
datascaled_new = scale_dataframe(data)

# Display the first few rows of the scaled DataFrame
datascaled_new.head()
```
